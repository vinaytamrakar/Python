```python
# preprocess and clean the content of a resume text before further analysis

# Import Regular Expressions.
import re
def cleanResume(resumeText):
    
    # remove punctuations
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resumeText)  
    
    # remove non-ascii characters
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText) 
    
    # remove extra whitespace
    resumeText = re.sub('\s+', ' ', resumeText)  
    
    # remove numbers
    resumeText = re.sub(r'[0-9]+', '', resumeText)  
    return resumeText.lower()
```

```python
# Creates a new column "Cleaned Resume" in the DataFrame containing the preprocessed text.
df["Cleaned Resume"] = df["Resume"].apply(lambda x: cleanResume(x))

# Length of the cleaned text in the first row of the "Cleaned Resume" column.
len(df["Cleaned Resume"][1])    # "len" used to measure the number of characters in the text
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk import word_tokenize

# calculates the number of stopwords in the English language
len(stopwords.words('english'))
```

```python
# take new DataFrame containing only rows where the "Category" column value is 'Data Science'.
ds_df = df[df.Category == 'Data Science']


#  Initializes an empty string
resumes=""

# Initializes an empty list
total_words = []

# iterates through each cleaned resume in the data science category
for resume in ds_df["Cleaned Resume"]:
    
    # Concatenates the cleaned resume to the resumes string.
    resumes += resume
    
    # Tokenizes the cleaned resume into a list of words
    words = word_tokenize(resume)
    
    # This condition checks if the word is not a stopword and not a punctuation character
    # If the word meets the condition, it's added to the total_words list.
    for word in words :
        if word not in set(stopwords.words('english')) and word not in string.punctuation:
            total_words.append(word)
```

```python
from sklearn.model_selection import train_test_split

# Extracting the "Cleaned Resume" text data and "Labels" from the DataFrame 'df'
text = df["Cleaned Resume"].values
labels = df["Labels"].values

#  split the data into training and testing sets
text_train,text_test,y_train,y_test = train_test_split(text, labels, random_state=0, test_size=0.25, stratify=df.Labels)
print(text_train.shape)
print(y_train.shape)
print(text_test.shape)
print(y_test.shape)
```

# Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
word_vectorizer = TfidfVectorizer(
sublinear_tf=True,
use_idf = True,
stop_words='english',
max_features=1000)
```

```python
X_train = word_vectorizer.fit_transform(text_train)
X_train.shape
```

```python
X_test = word_vectorizer.transform(text_test)
X_test.shape
```

```python
import itertools
dict(itertools.islice(word_vectorizer.vocabulary_.items(), 10))
```

# Multi-class Classification

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.multiclass import OneVsRestClassifier

# creates an instance of the OneVsRestClassifier and fits it with a MultinomialNB classifier
clf = OneVsRestClassifier(MultinomialNB()).fit(X_train, y_train)

# Making Predictions
prediction_mnb = clf.predict(X_test)

# calculates the accuracy of the classifier on the training set.
print('Accuracy of MultinomialNB Classifier on training set: {:.2f}'.format(clf.score(X_train, y_train)))

# calculates the accuracy of the classifier on the test set.
print('Accuracy of MultinomialNB Classifier on test set: {:.2f}'.format(clf.score(X_test, y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# OneVsRestClassifier allows to train a separate binary classification model for each class
# KNeighborsClassifier() is the base classifier that's being used within the OvR framework.
model = OneVsRestClassifier(KNeighborsClassifier()).fit(X_train, y_train)

# Making Predictions
prediction_knc = model.predict(X_test)
model.fit(X_train,y_train)

# calculates and prints the accuracy of the classifier on both the training and test sets
print('Accuracy of KNeighbors Classifier on training set: {:.2f}'.format(model.score(X_train, y_train)))
print('Accuracy of KNeighbors Classifier on test set: {:.2f}'.format(model.score(X_test, y_test)))
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, prediction_mnb))
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, prediction_knc))
```
